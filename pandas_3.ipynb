## NEW IPL DATASET

```python
delivery=pd.read_csv('deliveries.csv')
delivery.head(3)
```

```python
delivery.groupby('batsman')['batsman_runs'].count()
```

```python
death_over=delivery[delivery['over']>15]
death_over
```

```python
all_batsman=death_over.groupby('batsman')['batsman_runs'].count()
# HERE WE GET THE PLAYERS WHO HAVE PLAYED BALLS>200
```

```python
x=all_batsman>200
```

```python
x
```

```python
#HERE WE GET INDEXES OF ALL BATSMEN WHO HAVE PLAYED MORE THAN 200 BALLS
all_batsman[x].index
```

#### TO CONVERT A INDEX INTO A LIST WE HAVE tolist() FUNCTION

```python
batsman_list=all_batsman[x].index.tolist()
batsman_list
```

```python
delivery[delivery['batsman'].isin(batsman_list)]
# HERE WE ARE DIRECLTY SOTRING THE CHANGES IN THE DELIVERY DATASET
# THUS WE ONLY GET THE DATA IN WHICH THE ABOVE CONSTRAINT IS SATISFIED
```

```python
delivery[delivery['batsman'].isin(batsman_list)].shape
```

```python
final=delivery[delivery['batsman'].isin(batsman_list)]
```

```python
runs=final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls=final.groupby('batsman')['batsman_runs'].count()
```

```python
sr=(runs/balls)*100
```

```python
sr
```

## MERGE FUNCTION
### CALCULATING ORANGE CAP HOLDERS FROM THE DATA

```python
#MERGE FUNCTION CAN DO THE MERGING OF MULTIPLE DATA SETS
# THE ONLY CRITERIA HERE IS THERE SHOULD BE SOME SORT OF RELATION BETWEEN THE DATASETS
delivery.head(2)
```

```python
data.head(2)
```

#### BOTH THE DATAS ABOVE A COMMON COLUMN THAT IS id
##### THUS WE CAN USE THE merge FUNCTION HERE

```python
##HOW TO MERGE?
new=delivery.merge(data,left_on='match_id',right_on='id')
new
## HERE left_on TELLS THE LEFT DATA SET THAT IS 'deliveries.csv' AND 'right_on' TELLS THE RIGHT DATA SET THAT IS matches.csv
# WE HAVE TO MENTION THE COLUMNS WHICH REQUIRE TO COMBINE THE DATASETS
```

```python
new.groupby('batsman')['batsman_runs'].sum()
```

### TO CONVERT THE SERIES INTO DATAFRAME WE HAVE A FUNCTION CALLED RESET INDEX

```python
new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index()
#THUS HERE WE GET THE DATA SEASON WISE THAT HOW MANY RUNS EACH PLAYER HAS SCORED
```

# WE JUST THE WANT THE DATA OF HIGHEST RUN SCORER IN EACH SEASON

```python
new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season',keep='first')

```
