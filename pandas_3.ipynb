## sort_values function

```python
(data['team1'].value_counts()  + data['team2'].value_counts()).sort_values(ascending=False)
```

###  BUT CHANGE IN DATA IS NOT PERMENANT

```python
data.sort_values('city').head()  
```

```python
data.head()
```

# THUS TO PERMENANTLY SORT THE DATA WE HAVE A FUNCTION CALLED 'inplace'

```python
data.sort_values('city',inplace=True)
```

```python
data.head()
#WE CAN OBSERVE THE DATA GOT CHANGED PERMENANTLY
```

```python
data.sort_values('city',inplace=False)
```

```python
data.sort_values(['city','date'],ascending=[False,True]).head()
# city is not ascending and date is ascending by declaring a list
```

## drop_duplicates METHOD

```python
# HERE WE CAN EVEN USE INPLACE PARAMETER
data.drop_duplicates('city').head()
```

```python
data.drop_duplicates('city').head().shape
```

```python
data.drop_duplicates(['city','season']).head().shape
```

```python
data['winner'].head()
```

```python
data.drop_duplicates('season',keep='last')[['season','winner']].sort_values('season',ascending=True)
```
