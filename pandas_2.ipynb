```python
mask1=data['toss_winner']=='Royal Challengers Bangalore'
mask1
data[mask1]

#HERE WE ONLY GET WHERE ROYAL CHALLENGERS BANGLORE HAVE WON THE MATCH
#ALL THE FALSE CONDITION GETS REMOVED
```

```python
#function to display toss wins by any team
def toss(toss_winner):
    mask1=data['toss_winner']==toss_winner
    return data[mask1].shape[0]
    
```

```python
toss('Mumbai Indians')
```

```python
mask3=data['city']=='Pune'
mask3
```

```python
data[mask3].shape[0]
```

```python
#function to find matches played any particular city
def show_city(city):
    mask3=data['city']==city
    return mask3.shape[0]
```

```python
show_city('Rajkot')
```
