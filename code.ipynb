```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score

# True labels for ResNet (plant health)
true_plant_health_labels = ['Healthy', 'Unhealthy', 'Healthy', 'Healthy', 'Unhealthy']
# ResNet predictions
plant_health_predictions = ['Healthy', 'Unhealthy', 'Healthy', 'Unhealthy', 'Unhealthy']

# Calculate accuracy for ResNet
plant_health_accuracy = accuracy_score(true_plant_health_labels, plant_health_predictions)
print(f"Plant Health Model Accuracy: {plant_health_accuracy * 100:.2f}%")

# True labels for Random Forest (environmental risk)
true_env_risk_labels = ['Good', 'Poor', 'Moderate', 'Good', 'Poor']
# Random Forest predictions
env_risk_predictions = ['Good', 'Poor', 'Moderate', 'Good', 'Moderate']

# Calculate accuracy for Random Forest
env_risk_accuracy = accuracy_score(true_env_risk_labels, env_risk_predictions)
print(f"Environmental Risk Model Accuracy: {env_risk_accuracy * 100:.2f}%")

# True and predicted labels for the integrated model
true_recommendations = ['Healthy', 'Poor', 'Moderate', 'Healthy', 'Poor']  # Example data
generated_recommendations = ['Healthy', 'Poor', 'Moderate', 'Unhealthy', 'Poor']  # Example data

# Calculate accuracy for the integrated model
recommendation_accuracy = accuracy_score(true_recommendations, generated_recommendations)
print(f"Integrated Model Accuracy: {recommendation_accuracy * 100:.2f}%")

# Plotting the accuracies
models = ['Plant Health Model (ResNet)', 'Environmental Risk Model (RF)', 'Integrated Model']
accuracies = [
    plant_health_accuracy * 100,
    env_risk_accuracy * 100,
    recommendation_accuracy * 100
]

plt.figure(figsize=(10, 6))
plt.bar(models, accuracies, color=['blue', 'orange', 'green'])
plt.xlabel('Model')
plt.ylabel('Accuracy (%)')
plt.title('Model Accuracies')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add text annotations
for i, accuracy in enumerate(accuracies):
    plt.text(i, accuracy + 1, f'{accuracy:.2f}%', ha='center', va='bottom')

plt.show()
```
